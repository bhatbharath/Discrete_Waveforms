# 7. Program to generate Discrete waveforms


 ## a. Unit Step Function

```python
import numpy as np
import matplotlib.pyplot as plt

n=range(-1,6,1)
y=[]
for i in range(len(n)):
    temp=(1 if n[i]>=0 else 0)
    y.append(temp)
print(list(n))
print(y)

plt.stem(n,y)
plt.axis([-2,6,-0.1,1.5]) #Defines the range for the axis
plt.grid(True)
plt.show()
```

## b.Impulse Response

## c.Unit Ramp Function

```python
import numpy as np
import matplotlib.pyplot as plt

n=range(-1,6,1)
y=[]
for i in range(len(n)):
    temp=(n[i] if n[i]>=0 else 0)
    y.append(temp)
print(list(n))
print(y)

plt.stem(n,y)
plt.axis([-2,6,-1,6])
plt.grid(True)
plt.show()
```

## d.Real Exponential Function

### Decaying Exponential Function (0<a<1)

```python
import numpy as np
import matplotlib.pyplot as plt

a=0.9
n=range(0,6,1)
y=[]
for i in range(len(n)):
    temp=a**n[i]
    y.append(temp)
print(list(n))
print(y)

plt.stem(n,y)
plt.axis([-1,6,-1,2])
plt.grid(True)
plt.show()
```

### Growing Exponential Function (a>1)

```python
import numpy as np
import matplotlib.pyplot as plt

a=1.1
n=range(0,6,1)
y=[]
for i in range(len(n)):
    temp=a**n[i]
    y.append(temp)
print(list(n))
print(y)

plt.stem(n,y)
plt.axis([-1,6,-1,2])
plt.grid(True)
plt.show()
```

## e.Sine Function

```python
import numpy as np
import matplotlib.pyplot as plt

sr=100.0
ts=1.0/sr
t=np.arange(0,1,ts)
freq=5
y=np.sin(2*np.pi*freq*t)
print(list(y))

plt.stem(t,y)
plt.grid(True)
plt.show()
```

### Delay Shift of Ramp Function by 2 units

```python
import numpy as np
import matplotlib.pyplot as plt

n=range(-1,9,1)
y=[]
for i in range(len(n)):
    temp=(n[i-2] if n[i]>=2 else 0)
    y.append(temp)
print(list(n))
print(y)

plt.stem(n,y)
plt.axis([-2,10,-1,7])
plt.grid(True)
plt.show()
```
